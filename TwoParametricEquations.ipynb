```
import math
import matplotlib.pyplot as plt
#------------------------------------
def setup_T(min_x, max_x, dx):
    X = []; x = min_x
    while x <= max_x:             
        X.append(x)
        x += dx
    if x != max_x: X.append(max_x)
    return X
#------------------------------------
def plot(x, y, min_t, max_t, dt):
    T = setup_T(min_t, max_t, dt)
    X = [x(t) for t in T] 
    Y = [y(t) for t in T]
    plt.plot( X, Y, 'blue' )
    plt.show()
#====================================
```

```
# Prog-02: Beautiful Parametric Equations
# 2217365 สมชาย ประสิทธิ์จูตระกูล
# $x(t)=16sin^3(t)$
# $y(t)=13cos(t)-5cos(2t)-2cos(3t)-cos(4t)$
# โปรแกรมนี้แสดงสองฟังก์ชันที่ประกอบกันเป็นรูปหัวใจ เป็นฟังก์ชันแห่งความรักและปรองดอง
# ผมป็นผู้เลือกและเขียนนิพจน์คณิตศาสตร์ข้างบนนี้ด้วยตนเอง
def x(t):
    x1 = 16*math.sin(t)**3
    return x1

def y(t):
    y1 = 13*math.cos(t) - 5*math.cos(2*t) - 2*math.cos(3*t) - math.cos(4*t)
    return y1

plot(x, y, 0, 2*math.pi, 0.1)
```

```
# Prog-02: Beautiful Parametric Equations
# 2217365 สมชาย ประสิทธิ์จูตระกูล
# $x(t)=2.5sin(-5t)^22^{(cos(cos(9.844t)))}$  
# $y(t)=2.5sin(sin(-5t))cos(9.844t)^2$  
# โปรแกรมนี้แสดงสองฟังก์ชันที่ประกอบกันเป็นรูปอะไรสักอย่าง สวยดี
# ผมป็นผู้เลือกและเขียนนิพจน์คณิตศาสตร์ข้างบนนี้ด้วยตนเอง
def x(t):
    x1 = 2.5*math.sin(-5*t)**2 * 2**(math.cos(math.cos(4.28*2.3*t)))
    return x1

def y(t):
    y1 = 2.5*math.sin(math.sin(-5*t))*math.cos(4.28*2.3*t)**2
    return y1

plot(x, y, -6, 6, 0.01)
```
